```python
import numpy as np
import pandas as pd
```

Using scikit-learn (model, preprocessing, accuracy) :
  Scikit-learn used for its simpler and user friendly uses.

```python
# Using scikit-learn (model, preprocessing, accuracy)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

Using Scikit-learn dataset (iris) :
  

```python
# Using Scikit-learn dataset (iris)
from sklearn.datasets import load_iris
```

Using Scikit-learn's Support Vector Machine (SVM)

```python
# Using Scikit-learn's Support Vector Machine (SVM)
from sklearn.svm import SVC
```

```python
# Load Iris Dataset
iris = load_iris()
```

```python
# Features (data) and Labels (targets)
x = iris.data   # Features
y = iris.target # Target labels (species)
```

```python
import matplotlib.pyplot as plt

# Scatter plot of first two features
plt.scatter(x[:, 0], x[:, 1], c=y)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Iris Dataset: Sepal Length vs Sepal Width")
plt.show()
```

```python
# Split data into training and testing (80% training and 20% testing)
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.2,
    random_state = 42
)
```

```python
# Standardize Features using StandardScaller (( Preprocessing ))
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Initialize the model
model = SVC()
```

```python
# Train the model
model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(x_test)
```

```python
# Calculate the Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy : {accuracy * 100:.2f}%")
```

```python
# Classification Test
sample_input = np.array([[5.1, 3.5, 1.4, 0.2]])
# Apply the same scaler used for training
sample_input_scaled = scaler.transform(sample_input)


# Predict the class
predicted_class = model.predict(sample_input_scaled)[0]

print("Predicted class : ", iris.target_names[predicted_class])
```
